```python
from matplotlib import pylab as plt
import numpy as np
import math
%matplotlib inline
```

```python
Im = float(input("Введите амплитудное значение тока: "))
f0 = float(input("Введите начальную фазу: "))
w = 5
t = [i for i in range(100)]
I = [(Im*math.sin(w*i+f0)) for i in range(100)]
plt.plot(t,I)
```
